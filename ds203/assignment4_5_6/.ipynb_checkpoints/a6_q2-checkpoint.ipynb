```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('q2.csv')
```

```python
# preprocessing size
def process_size(s):
    s=str(s)
    s=s.lower()
    s = s.replace(" ", "")
    s = s.replace("size", "")
    s = s.replace(".",  "")
    s = s.replace("-", "")
    try:
        s_0 = int(s[0])
        if s[1]=="x":
            s_pre = "x"*(s_0-1)
            s = s_pre + s[1:]
    except:
        pass
    return s
df["product_variation_size_id"]=df["product_variation_size_id"].apply(process_size)
vals, counts = np.unique(np.array(df["product_variation_size_id"], dtype=str), return_counts=True)
vals = vals[counts.argsort()[::-1]] # most frequent vals
def reduce_sizes(size):
    # if it is a frequent size, then keep
    if size in vals[:10]:
        return size
    # else categorize as others
    return "nan"
df["product_variation_size_id"]=df["product_variation_size_id"].apply(reduce_sizes)
```

```python
# preprocessing product_color
vals, counts = np.unique(np.array(df["product_color"], dtype=str), return_counts=True)
vals = vals[counts.argsort()[::-1]] # most frequent vals
def reduce_colors(color):
    # if it is a frequent color, then keep
    if color in vals[:20]:
        return color
    # else categorize as others
    return "other"
df["product_color"]=df["product_color"].apply(reduce_colors)
```

```python
vals, counts = np.unique(np.array(df["product_variation_size_id"], dtype=str), return_counts=True)
vals = vals[counts.argsort()[::-1]] # most frequent sizes
top_sizes=vals[:6]
top_sizes
```

```python
df_new = df.loc[df["product_variation_size_id"].isin(top_sizes)]
```

```python
print("Size Mean Variance")
for s in top_sizes:
    print(s, round(df_new[df_new["product_variation_size_id"]==s]["units_sold"].mean(), 3),
       round(df_new[df_new["product_variation_size_id"]==s]["units_sold"].var(), 3))
```

```python
from scipy.stats import gamma
import statsmodels.api as sm 
plt.style.use("seaborn")

for s in top_sizes:
    sm.qqplot(data=df_new[df_new["product_variation_size_id"]==s]["units_sold"], line='s')
    plt.title(s)
    plt.show()
```

```python
def ll_wrt_norm(x, mu, sigma):
    normalization = 1/(((2*np.pi)**0.5)*sigma)
    p = (-((x-mu)**2)/(2*sigma**2))
    return np.log(normalization) + p
def ll_for_whole(vals, mu, sigma):
    in_lls = ll_wrt_norm(vals, mu, sigma)
    return np.sum(in_lls)
```

```python
print("Log likelihoods of units_sold wrt normal distribution: ")

for s in top_sizes:
    cnt = 1
    mu = df_new[df_new["product_variation_size_id"]==s]["units_sold"].mean()
    sigma = df_new[df_new["product_variation_size_id"]==s]["units_sold"].var()**0.5
    ans = ll_for_whole(df_new[df_new["product_variation_size_id"]==s]["units_sold"], mu, sigma)
    print(s,round(ans, 3))
```

The closer the points lie to the line,  the higher the log likelihood. i.e. If the straight line fits well, the likelihood is higher
In this case, the gaussian distribution gives very low log likelihood

```python
import scipy.stats as stats

print("95% CI - Z interval, T interval : ")
for s in top_sizes:
    mu = df_new[df_new["product_variation_size_id"]==s]["units_sold"].mean()
    sigma = np.sqrt(df_new[df_new["product_variation_size_id"]==s]["units_sold"].var())
    tot = len(list(df_new[df_new["product_variation_size_id"]==s]))
    z_lower = mu - stats.norm.ppf(0.975)*(sigma/(np.sqrt(tot)))
    z_upper = mu + stats.norm.ppf(0.975)*(sigma/(np.sqrt(tot)))
    z_interval = (round(z_lower, 3), round(z_upper, 3))
    tot_sui = df_new[df_new["product_variation_size_id"]==s]["units_sold"]
    t_interval = stats.t.interval(0.95, len(tot_sui)-1, scale=stats.sem(tot_sui), loc=np.mean(tot_sui))
    t_interval = (round(t_interval[0], 3), round(t_interval[1], 3))
    print(s, z_interval, t_interval)
```

```python
# Closest means are S and L
s1="s"
s2="l"
d1=df_new[df_new["product_variation_size_id"]==s1]["units_sold"]
d2=df_new[df_new["product_variation_size_id"]==s2]["units_sold"]
mu1 = d1.mean()
mu2 = d2.mean()
var1 = d1.var()
var2 = d2.var()
n1 = d1.shape[0]
n2 = d2.shape[0]
t = (mu1 - mu2) / np.sqrt((var1/n1) + (var2/n2))

#compute degrees of freedom
DoF = np.square((var1/n1) + (var2/n2)) / (((var1/n1)**2)/(n1-1) + ((var2/n2)**2)/(n2-1))

#compare values
threshold=0.05
critical_val = stats.t.pdf((1-threshold)/2,DoF)
print(np.abs(t), critical_val)
if np.abs(t) > critical_val:
    print('Null hypothesis is rejected')
    print('YES')
else:
    print('Null hypothesis is not rejected')
    print('NO')
```

```python
# l has fewer items, so we randomly sample items from s
d1_ = d1.sample(frac = d2.shape[0]/d1.shape[0])
```

```python
#signed rank test
from scipy.stats import wilcoxon

w,p = wilcoxon(d1_.to_numpy()-d2.to_numpy())
print('w = '+str(w)+', p = '+str(p))

if p<=threshold:
    print('Null hypothesis is rejected')
else:
    print('Not enough evidence to reject')
```

```python
t = stats.t.pdf((1-threshold)/2,n2-1)

std = np.std(d1_.to_numpy()-d2.to_numpy())
t_interval =((mu1-mu2) - t*std/np.sqrt(n2),(mu1-mu2) + t*std/np.sqrt(n2))

# does 0 belong to interval?
if 0>t_interval[0] and 0<t_interval[1]:
    print("Insufficient evidence")
else:
    print(t_interval)
    print("Null hypothesis rejected since zero does not lie in interval")
```
