```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("q1.csv")
```

```python
tot_suicides=df.groupby(by=["country"])["suicides_no"].sum()
```

```python
tot_pop = df.groupby(by=["country"])["population"].sum()
```

```python
avg = tot_suicides / tot_pop
```

```python
top_countries = avg.sort_values(ascending=False)[:6].keys()
```

```python
top_countries
```

```python
tot_suicides_yearly = df.groupby(by=["country", "year"])["suicides_no"].sum()
```

```python
for c in top_countries:
    data = tot_suicides_yearly.loc[c]
    x=data.index.get_level_values('year')
    sns.barplot(y=data.values, x=x, palette="rainbow").set_title('Total annual suicides ')
    plt.xticks(rotation="70")
    plt.show()
```
