```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
font = {'family' : 'Arial',
       'weight': 'light',
       'size': 11}

plt.rc('font', **font)

SMALL_SIZE = 11
MEDIUM_SIZE = 15
BIGGER_SIZE = 18
plt.rc('axes', titlesize=BIGGER_SIZE)
plt.rc('axes', labelsize=MEDIUM_SIZE)
```

```python
df = pd.read_csv("q2.csv")
```

```python
df.columns
```

```python
# preprocessing size
def process_size(s):
    s=str(s)
    s=s.lower()
    s = s.replace(" ", "")
    s = s.replace("size", "")
    s = s.replace(".",  "")
    s = s.replace("-", "")
    try:
        s_0 = int(s[0])
        if s[1]=="x":
            s_pre = "x"*(s_0-1)
            s = s_pre + s[1:]
    except:
        pass
    return s
df["product_variation_size_id"]=df["product_variation_size_id"].apply(process_size)
vals, counts = np.unique(np.array(df["product_variation_size_id"], dtype=str), return_counts=True)
vals = vals[counts.argsort()[::-1]] # most frequent vals
def reduce_sizes(size):
    # if it is a frequent size, then keep
    if size in vals[:10]:
        return size
    # else categorize as others
    return "nan"
df["product_variation_size_id"]=df["product_variation_size_id"].apply(reduce_sizes)
```

```python
# preprocessing product_color
vals, counts = np.unique(np.array(df["product_color"], dtype=str), return_counts=True)
vals = vals[counts.argsort()[::-1]] # most frequent vals
def reduce_colors(color):
    # if it is a frequent color, then keep
    if color in vals[:20]:
        return color
    # else categorize as others
    return "other"
df["product_color"]=df["product_color"].apply(reduce_colors)
```

```python
# categorizing items based on price
df["price"].plot.hist()
plt.show()
def categorize_price(p):
    if p<5:
        return "Very Low"
    if p<10:
        return "Low"
    if p<15:
        return "High"
    return "Very High"
df["price_bins"]=df["price"].apply(categorize_price)
```

```python
data = df.groupby("product_variation_size_id")["units_sold"].sum()
ax=pd.DataFrame(data.values, index=data.index.get_level_values("product_variation_size_id")).plot.pie(
    y=0, labeldistance=None)
ax.set_title("Size wise distribution of sales")
ax.set(ylabel="Total Sales")
ax.tick_params(axis='y', pad=20)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

data = df.groupby("product_variation_size_id")["units_sold"].mean()
ax=pd.DataFrame(data.values, index=data.index.get_level_values("product_variation_size_id")).plot.pie(
    y=0, labeldistance=None)
ax.set_title("Mean sales for different sizes")
ax.set(ylabel="Avg. Sales")
ax.tick_params(axis='y', pad=20)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# s size has a surprisingly high contribution to total sales
# but avg. sales for size s is much smaller
# m and xs are other highly sold sizes
# xl and m have a high mean sale
```

```python
ax = sns.catplot(x="price_bins", y="units_sold", hue="uses_ad_boosts", kind = 'bar', data=df, palette = "rainbow")
axes = ax.axes.flatten()
axes[0].set_title("Seeing effect of ad boosts over various price ranges", fontsize=BIGGER_SIZE+7)
axes[0].set_ylabel("Total Sales")
axes[0].set_xlabel("Price Groups")

plt.xticks(rotation="70")
plt.rc('axes', titlesize=BIGGER_SIZE)
plt.rc('axes', labelsize=MEDIUM_SIZE)

plt.setp(ax._legend.get_texts(), fontsize=BIGGER_SIZE) # for legend text
ax._legend.set_title("Ad Boosts Used")
plt.setp(ax._legend.get_title(), fontsize=BIGGER_SIZE+3) # for legend title

new_labels = ['No', 'Yes']
for t, l in zip(ax._legend.texts, new_labels): t.set_text(l)

ax._legend.set_bbox_to_anchor((1.3, 0.9))
plt.show()
```

```python
# ad_boosts help only when prices are higher
# and in fact spoil sales for lower prices
```

```python
import matplotlib as mpl
```

```python
# units sold vs color with each bar representing the color

new_df2 = df.groupby("product_color")["units_sold"].mean()
new_df1 = df.groupby("product_color")["units_sold"].sum()
pal={}
predefined_cols={
    "navyblue": (0,0,0.5,1),
    "armygreen": (0.29, 0.33, 0.13, 1),
    "other": (0.6,0.6,0.6,0.5),
    "winered": (0.69, 0.07, 0.14, 1),
    "multicolor": (0,0,0,0.7),
    "white & green": (0,0.5,0,1)
}
for color in new_df1.index:
    try:
        pal[color]=mpl.colors.to_rgba(color)
    except:
        pal[color]=predefined_cols[color]
g = sns.barplot(x=new_df2.index, y=new_df2.values, palette=pal, linewidth=1, edgecolor=".2")
plt.xticks(rotation="70")
g.set_title("Units sold vs Color")
g.set(xlabel="Color", ylabel="Avg. units sold")
plt.show()

g = sns.barplot(x=new_df1.index, y=new_df1.values, palette=pal, linewidth=1, edgecolor=".2")
plt.xticks(rotation="70")
g.set_title("Which color sells the most?")
g.set(xlabel="Color", ylabel="Total units sold")
plt.show()
```
