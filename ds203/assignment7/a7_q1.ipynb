```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import sklearn.model_selection as ms
```

```python
df = pd.read_excel("ENB2012_data.xlsx")
```

```python
df
```

```python
# checking for missing data
df.isnull().sum()
```

```python
# separating Y
df_y1 = df["Y1"]
df_y2 = df["Y2"]
```

```python
# dropping Y columns
df.drop(["Y1", "Y2"], axis=1, inplace=True)
```

```python
# split into train and test sets
train_size=0.8
X_train1, X_test1, y_train1, y_test1 = ms.train_test_split(df.to_numpy(), df_y1.to_numpy(), train_size=train_size)
X_train2, X_test2, y_train2, y_test2 = ms.train_test_split(df.to_numpy(), df_y2.to_numpy(), train_size=train_size)
```

```python
# fit normalizer on training set and use it to transform test set
normalizer1 = StandardScaler()  
X_train_norm1 = normalizer1.fit_transform(X_train1)
X_test_norm1 = normalizer1.transform(X_test1)
normalizer2 = StandardScaler()  
X_train_norm2 = normalizer2.fit_transform(X_train2)
X_test_norm2 = normalizer2.transform(X_test2)
```

```python
import sklearn.linear_model as lm
from sklearn.metrics import r2_score, mean_squared_error
```

```python
lin_reg1 = lm.LinearRegression()
lin_reg1.fit(X_train_norm1, y_train1)
print("Coefficients:", lin_reg1.coef_)
print("Intercept:", lin_reg1.intercept_)
```

```python
lin_reg2 = lm.LinearRegression()
lin_reg2.fit(X_train_norm2, y_train2)
print("Coefficients:", lin_reg2.coef_)
print("Intercept:", lin_reg2.intercept_)
```

```python
y_pred1 = lin_reg1.predict(X_test_norm1)
print(y_pred1)
```

```python
y_pred2 = lin_reg2.predict(X_test_norm2)
print(y_pred2)
```

```python
print("R2 score:", r2_score(y_test1, y_pred1), "MSE:", mean_squared_error(y_test1, y_pred1))
```

```python
print("R2 score:", r2_score(y_test2, y_pred2), "MSE:", mean_squared_error(y_test2, y_pred2))
```

```python
# combining into one function
def train_lin_reg(df, df_y, train_size):
    X_train, X_test, y_train, y_test = ms.train_test_split(df.to_numpy(), df_y.to_numpy(), train_size=train_size)
    normalizer = StandardScaler()  
    X_train_norm = normalizer.fit_transform(X_train)
    X_test_norm = normalizer.transform(X_test)
    lin_reg = lm.LinearRegression()
    lin_reg.fit(X_train_norm, y_train)
    y_pred = lin_reg.predict(X_test_norm)
    print("R2 score:", r2_score(y_test, y_pred), "MSE:", mean_squared_error(y_test, y_pred))
```

```python
print("Y1: Heating Load")
train_lin_reg(df, df_y1, train_size=0.8)
print("Y2: Cooling Load")
train_lin_reg(df, df_y2, train_size=0.8)
```

```python
train_sizes=[0.6, 0.7, 0.8, 0.9]
for ts in train_sizes:
    print(ts, ":")
    print("Y1: Heating Load")
    train_lin_reg(df, df_y1, train_size=ts)
    print("Y2: Cooling Load")
    train_lin_reg(df, df_y2, train_size=ts)
    print()
```

```python
# 0.6 is suitable for Y1
# 0.8 is suitable for Y2
```

```python
def train_ridge_reg(df, df_y, lam, train_size=0.8):
    X_train, X_test, y_train, y_test = ms.train_test_split(df.to_numpy(), df_y.to_numpy(), train_size=train_size)
    normalizer = StandardScaler()  
    X_train_norm = normalizer.fit_transform(X_train)
    X_test_norm = normalizer.transform(X_test)
    ridge_reg = lm.Ridge(alpha = lam)
    ridge_reg.fit(X_train_norm, y_train)
    y_pred = ridge_reg.predict(X_test_norm)
    print("R2 score:", r2_score(y_test, y_pred), "MSE:", mean_squared_error(y_test, y_pred))
```

```python
lam_vals= [0.001, 0.005, 0.01, 0.05, 0.1, 0.5]
for ts in train_sizes:
    for lam in lam_vals:
        print(lam, ts)
        print("Y1: Heating Load")
        train_ridge_reg(df, df_y1, lam, ts)
        print("Y2: Cooling Load")
        train_ridge_reg(df, df_y2, lam, ts)
        print()
```

```python
def train_lasso_reg(df, df_y, lam, train_size=0.8):
    X_train, X_test, y_train, y_test = ms.train_test_split(df.to_numpy(), df_y.to_numpy(), train_size=train_size)
    normalizer = StandardScaler()  
    X_train_norm = normalizer.fit_transform(X_train)
    X_test_norm = normalizer.transform(X_test)
    lasso_reg = lm.Lasso(alpha = lam, max_iter=5000)
    lasso_reg.fit(X_train_norm, y_train)
    y_pred = lasso_reg.predict(X_test_norm)
    print("R2 score:", r2_score(y_test, y_pred), "MSE:", mean_squared_error(y_test, y_pred))
```

```python
lam_vals= [0.001, 0.005, 0.01, 0.05, 0.1, 0.5]
for ts in train_sizes:
    for lam in lam_vals:
        print(lam, ts)
        print("Y1: Heating Load")
        train_lasso_reg(df, df_y1, lam, ts)
        print("Y2: Cooling Load")
        train_lasso_reg(df, df_y2, lam, ts)
        print()
```
